### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
#Removing null values
event_date['Total.Fatal.Injuries']= event_date['Total.Fatal.Injuries'].fillna(0)
event_date['Number.of.Engines']=  event_date['Number.of.Engines'].fillna(0)
event_date['Total.Minor.Injuries']= event_date['Total.Minor.Injuries'].fillna(0)
event_date['Total.Uninjured']= event_date['Total.Uninjured'].fillna(0)
event_date['Total.Serious.Injuries']= event_date['Total.Serious.Injuries'].fillna(0)
event_date['Aircraft.damage'] = event_date['Aircraft.damage'].fillna('unknown')
```

```python
#Finding the total passengers
event_date['Total.Passengers'] = (
    event_date['Total.Fatal.Injuries'] +
    event_date['Total.Serious.Injuries'] +
    event_date['Total.Minor.Injuries'] +
    event_date['Total.Uninjured'])

event_date['Total.Passengers']= event_date['Total.Passengers'].fillna(0)
```

```python
#Finding the  % of total fatal injuries
event_date['% Fatal.Injuries'] =(event_date['Total.Fatal.Injuries']/ event_date['Total.Passengers']) * 100
event_date['% Serious.Injuries'] =(event_date['Total.Serious.Injuries']/event_date['Total.Passengers']) *100
event_date['% Minor.Injuries'] =(event_date['Total.Minor.Injuries']/event_date['Total.Passengers']) *100
event_date['% Serious/Fatal.Injured'] =((event_date['Total.Fatal.Injuries']+event_date['Total.Serious.Injuries'])/event_date['Total.Passengers']) * 100
event_date
```

Rows with 0 total passengers likely have missing injury data.
I treated these as 0% injury rate

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

These is inconsistency in naming the airplanes resulting in more entries of the makes than they actully are. For example the 'air tractor inc' and 'air tractor' also BOEING and boeing.
        
         Cleaning Tasks
1. Standardize text (lowercase → title case)
2. Strip whitespace
3. Normalize manufacturer names
4. Handle missing values
5. Filter out rare manufacturers (threshold ≥ 50 records)

```python
#Standardizing text
event_date['Make'] = event_date['Make'].str.strip().str.lower()
```

```python
#Removing NaNs
event_date = event_date.dropna(subset=['Make'])
```

```python
# Normalizing the naming of the manufacturers
make_mapping = {
    'cirrus design corp': 'cirrus',
    'air tractor inc': 'air tractor',     
    'grumman acft eng cor-schweizer': 'grumman',
    'aviat aircraft inc': 'aviat',
    'bombardier inc':'bombardier',
    'dehavilland': 'de havilland',
    'american champion aircraft': 'champion'}


event_date['Make'] = event_date['Make'].replace(make_mapping)  # Add mapping to the 'Make' column

```

```python
# Filtering out rare manufacturers
make_counts = event_date['Make'].value_counts()
valid_makes = make_counts[make_counts >= 50].index

event_date = event_date[event_date['Make'].isin(valid_makes)]
```
